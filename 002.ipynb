```python
def plot_separator(model,xs,ys,label = ' ',ax = None):
    ''' xs,ys,are 1-D b/c contour and decision_function use incompatible packaging'''
    if ax is None :
        ax = plt.gca()
    xy = np_cartesian_product(xs,ys)
    z_shape = (xs.size , ys.size)
    zs = model.decision_function(xy),reshape(z_shape)
    contours = ax.contour(xs,ys,zs,colors= 'k' , levels = [0],linestyles=['-'])
    fmt = {contours.levels[0]: label }
    labels = ax.clabel(contours, fmt=fmt , inline_spacing =10)
```

```python
number_people = np.arange(1,11)
number_rbs = np.arange(0,20)

#numpy tool to get cross product of values in two paired ways 
#try it out 

number_people , number_rbs = np.meshgrid(number_people , number_rbs)
total_cost = 80 * number_people + 10* number_rbs + 40
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig,axes = plt.subplots(2,3,subplot_kw ={'projection' : '3d'},figsize =(9,6))

angles = [0,45,90,135,180]
for ax,angle in zip(axes.flat,angles):
    ax.plot_surface(number_people ,number_rbs , total_cost)
    ax.set_xlabel("people")
    ax.set_ylabel("RootBeers")
    ax.set_zlabel("TotalCost")
    ax.azim = angle
    #we dont use the last axis
    axes.flat[-1].axis('off')
    fig.tight_layout()
```

Breaking Straight - jacket , and nonlinearity

```python
fig,axes = plt.subplots(2,2)
fig.tight_layout()

titles = [ "$y = c_0$", "$y = c_1x+c_0$" , "$y = c_2x^2 + c_1x + c_0$" , "$y = c_3x^3+c_2x^2+c_1x+c_0$"]
xs = np.linspace(-10 , 10 ,100)
for power , (ax,title) in enumerate (zip(axes.flat,titles),1):
    coeffs = np.random.uniform(-5,5,power)
    poly = np.poly1d(coeffs)
    ax.plot(xs,poly(xs))
    ax.set_title(title)
```
